```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

#Grafo que representa a rede de amizades de 34 pessoas em uma academia de karate em 1970
import networkx as nx
import matplotlib.pyplot as plt
import math, os, sys
from scipy.stats import cumfreq
import numpy as np
import statsmodels.api as sm
from statsmodels.distributions.empirical_distribution import ECDF
G=nx.Graph()
G = nx.read_gml('C:\\Users\\rapha\\Desktop\\Work\\Mestrado\\Redes Complexas\\karate.gml')
nx.draw(G)
plt.show()

#Closeness
closeness = (nx.closeness_centrality(G,u=None,distance=None, normalized=True))
avgCloseness = sum(nx.closeness_centrality(G).values())/G.number_of_nodes()
varClose = 0
for close in closeness:
    varClose += (close - avgCloseness) ** 2

varClose = varClose/G.number_of_nodes()
desvioClose = round(math.sqrt(varClose),2)

print 'sorted',sorted(nx.closeness_centrality(G).values(), reverse=True)
print 'Menos centrais: ',sorted(nx.closeness_centrality(G).values())
print 'Centralidade Closeness: ', closeness
print 'Centralidade de Closeness Mínima: ',min(nx.closeness_centrality(G).values())
print 'Centralidade de Closeness Máxima: ',max(nx.closeness_centrality(G).values())
print 'Centralidade de Closeness Média: ', avgCloseness
print 'Desvio Padrão de Closeness: ', desvioClose
print ""

#Betweenness
betweenness = nx.betweenness_centrality(G, k=None, normalized=True, weight=None, endpoints=False, seed=None)
avgBetweenness = sum(nx.betweenness_centrality(G).values())/G.number_of_nodes()
varBetween = 0
for between in betweenness:
    varBetween += (between - avgBetweenness) ** 2

varBetween = varBetween/G.number_of_nodes()
desvioBetween = round(math.sqrt(varBetween),2)

print 'sorted',sorted(nx.betweenness_centrality(G).values(), reverse=True)
print 'Menos centrais: ',sorted(nx.betweenness_centrality(G).values())
print 'Centralidade de Betweenness: ', betweenness
print 'Centralidade de Betweenness Mínima: ', min(nx.betweenness_centrality(G).values())
print 'Centralidade de Betweenness Máxima: ', max(nx.betweenness_centrality(G).values())
print 'test'
print 'Centralidade de Betweenness Média: ', avgBetweenness
print 'Desvio Padrão de Betweenness: ', desvioBetween
print ''

#Katz
katz = nx.katz_centrality(G, alpha=0.1, beta=1.0, max_iter=1000, tol=1e-06, nstart=None, normalized=True, weight='weight')
avgKatz = sum(nx.katz_centrality(G).values())/G.number_of_nodes()
varKatz = 0
for ktz in katz:
    varKatz += (ktz - avgKatz) ** 2

varKatz = varKatz/G.number_of_nodes()
desvioKatz = round(math.sqrt(varKatz),2)

print 'Mais centrais: ',sorted(nx.katz_centrality(G).values(), reverse=True)
print 'Menos centrais: ',sorted(nx.katz_centrality(G).values())
print 'Centralidade de Katz: ', katz
print 'Centralidade de Katz Mínima: ', min(nx.katz_centrality(G).values())
print 'Centralidade de Katz Máxima: ', max(nx.katz_centrality(G).values())
print 'Centralidade de Katz Média: ', avgKatz
print 'Desvio Padrão de Katz: ', desvioKatz
print ''

pr = nx.pagerank(G, alpha=0.9)
print "Centralidade de PageRank: ", pr
print 'sorted',sorted(nx.pagerank(G, alpha=0.9).values(), reverse=True)
```
